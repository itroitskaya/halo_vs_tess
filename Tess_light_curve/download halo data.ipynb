```python
import requests, re
from os import path

page_url = "https://archive.stsci.edu/hlsp/halo#section-2bc7ad1a-7a8c-445e-853f-b09ff1364a13"
data_dir = "fits"

print("Getting page... ", end='')
page_content = requests.get(page_url, allow_redirects=True).content
print("ok")

page_str = str(page_content)

search_r = r'href="(https[^"]*?_lc.fits)"'
name_r = r'.*/([^/]*\d{9})'

links = re.findall(search_r, page_str)
links_count = len(links)
print('Downloading {} files'.format(links_count))

for i, l in enumerate(links):
    print('{:3d}/{:<3d} {}\t'.format(i + 1, links_count, l), end='')
    match = re.search(name_r, l)
    file_name = path.realpath(path.join(data_dir, match.group(1) + '_lc.fits'))
    file_content = requests.get(l).content
    with open(file_name,"wb") as f:
        f.write(file_content)
    print("ok")
```
